```kotlin
// Ensures that the latest available library versions are used
%useLatestDescriptors

// Imports the Kotlin DataFrame library
%use dataframe

// Imports the Kotlin Kandy library
%use kandy
```

```kotlin
val rawDf = DataFrame.read("/Users/sokolova/jacoco/all_modules_report_3410.csv")
rawDf
```

```kotlin
val dbWithcoverage = rawDf.add("INSTRUCTION_COVERAGE") {
    if (INSTRUCTION_COVERED + INSTRUCTION_MISSED != 0) {
        INSTRUCTION_COVERED.toDouble().div(INSTRUCTION_COVERED + INSTRUCTION_MISSED)
    } else {
        0.0
    }
}
dbWithcoverage
```

```kotlin
import org.jetbrains.kotlinx.statistics.dataframe.stat.mean

val meanCoverage = dbWithcoverage.groupBy("GROUP").mean("INSTRUCTION_COVERAGE").sortBy("GROUP")
meanCoverage
```

```kotlin
meanCoverage.plot {
    bars {
        x(GROUP)
        y(INSTRUCTION_COVERAGE)
    }
    layout {
        title = "Instruction coverage per module"
    }
}
```
